```python
#Q2
import pandas as pd

df1 = pd.DataFrame({
    "Name":["cherry","mango","potato","orion"],
    "Type":["Fruit","Fruit","Vegetable","Vegetable"],
    "Price":[100, 110, 60, 80]
})

df2 = pd.DataFrame({
    "Name":["peper","carrot","banana","kiwi"],
    "Type":["Vegetable","Vegetable","Fruit","Fruit"],
    "Price":[50, 70, 90, 120]
})

df3 = pd.concat([df1, df2])

df_fruit = df3.sort_values(["Type","Price"],ascending=[False,False])
df_veg = df3.sort_values(["Type","Price"],ascending=[True,False])

print("Sum of Top2 Fruits Price:", sum(df_veg.Price[:2]))
print("Sum of Top2 Vegtable Price:", sum(df_fruit.Price[:2]))
```

```python
#Q3
import pandas as pd

idx = ["Sue","Ryan","Jay","Jane","Anna"]
col = ["round_1","round_2","round_3","round_4","round_5"]
data = [[55, 65, 60, 66, 57],
        [64, 77, 71, 79, 67],
        [88, 81, 79, 89, 77],
        [45, 35, 38, 46, 47],
        [91, 96, 90, 97, 99]]

df = pd.DataFrame(data, columns = col, index=idx)

col_round_6 = [11, 15, 13, 17, 19]
df['round_6'] =col_round_6
print(df)

data2 = [[df[i].mean() for i in df],
        [df[i].max() for i in df],
        [df[i].min() for i in df]]

df2 = pd.DataFrame(data2, columns = df.columns, index =['mean','max','min'])

print(df2)
```

```python
#Q4
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0., 5., 0.2)
Y = [value**3 for value in t]
W = [i**2 for i in t]
Z = [i for i in t]

plt.plot( t, Z, color = 'red', linestyle ="dashed")
plt.scatter( t, W, color = 'green' , marker="s")
plt.scatter( t, Y, color = 'blue', marker="^"  )
plt.show()
```
