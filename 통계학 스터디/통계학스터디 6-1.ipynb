```python
# 6 일반선형모델
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats

from matplotlib import pyplot as plt
import seaborn as sns
sns.set()

import statsmodels.formula.api as smf
import statsmodels.api as sm

%precision 10

%matplotlib inline
```

```python
# 6.1.6 이항분포
sp.stats.binom.pmf( k = 1, n = 2,p = 0.5) # pmk Probability Mass function 확률질량함수
```

```python
np.random.seed(1)
sp.stats.binom.rvs( n = 10, p = 0.2, size =5) # rvs random variates 변수 변량 variable 과 비슷
```

```python
binomial = sp.stats.binom(n = 10, p = 0.2)

np.random.seed(1)
rvs_binomial = binomial.rvs(size = 10000)

m = np.arange(0,10,1)
pmf_binomial = binomial.pmf( k = m)
sns.distplot(rvs_binomial, bins = m,kde=False,norm_hist= True) #norm_hist 없으면 겹쳐그리기 안됨
plt.plot(m, pmf_binomial, color = "black")
```

```python
# 6.1.7 푸아송분포

sp.stats.poisson.pmf(k=2, mu = 5)
```
