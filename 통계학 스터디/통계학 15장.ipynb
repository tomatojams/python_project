```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats

from matplotlib import pyplot as plt

import seaborn as sns

sns.set()
%precision 10
%matplotlib inline
```

```python
x_plot = np.arange(start = -1.265, stop  = 2.065, step = 0.001)
plt.plot(x_plot, stats.norm.pdf(x = x_plot, loc= 0, scale = 1/np.sqrt(10)), color ='black')
```

```python
x_plot = np.arange(start = -1.265, stop  = 2.065, step = 0.001)
plt.plot(x_plot, stats.norm.pdf(x = x_plot, loc= 0.8, scale = 1/np.sqrt(10)), color ='black')
```
