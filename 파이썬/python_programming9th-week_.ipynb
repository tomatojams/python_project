**함수의 정의**

```python
def add(a,b):        # a, b는 매개 변수
    return a+b      
```

**함수의 호출**

```python
c = add(3,4)      # 위에서 정의한 add 함수의 호출    3,4 는 인수

print(c)
```

**입력값이 없는 함수**

```python
def say():
    return 'HI'

a = say()    # 함수를 정의할때 입력값이 없으므로, say() 함수 호출시에도 인수를 넣지 않아야 함

print(a)
```

**결과값이 없는 함수**

```python
def add(a,b):
    print("%d, %d의 합은 %d입니다."%(a,b,a+b))    # return 이 없다

add(3,4)
```

**함수의 초깃값을 미리 설정**

```python
def say_myself(name,old,man=True):
    print("나의 이름은 %s 입니다."%name)
    print("나의 나이는 %d 입니다."%old)
    if man:
        print("남자 입니다.")
    else:
        print("성별을 알 수없습니다.") 

say_myself("박응용",27)      # 세번째 인자에 값을 넣지 않으면 초기값으로 설정된 True가 man에 들어감
 
say_myself("박응용",27,False)  # 세번째 인자에 값을 넣지 않으면 초기값으로 설정된 True가 man에 들어감
```

**지역 변수**

```python
a = 1
def vartest(a):
    a = a + 1

vartest(a)         # 함수를 호출해서 a 값을 전달했고, 분명 함수 안에서 a 에 1을 더했는데.......
print(a)          # a의 값을 살펴보면 1이 출력된다.    이것은 함수 안의 a는 함수 안에서만 사용하는 지역변수 a 이기 때문이다.
```

**그렇다면 위의 코드에 a에 1이 더해진 결과를 얻기 위해서는 어떻게 해야 할까?**

```python
a = 1
def vartest(a):
    return a+1

a=  vartest(a)
print(a)          # a의 값을 살펴보면 1이 출력된다.    이것은 함수 안의 a는 함수 안에서만 사용하는 지역변수 a 이기 때문이다.
```

**lambda 함수의 사용**

**readline() 함수 이용하기**

```python
# readline_test.py
f = open("python.txt", 'r')
line = f.readline()      # readline 함수는 한 줄만 읽는다.
print(line)
f.close()
```

**반복문을 이용해 모든 값 읽어오기**

```python
f = open("python.txt", 'r')
while True:
    line = f.readline()
    if not line: break
    print(line)
f.close()
```

**readlines 함수 사용하기**

```python
f = open("python.txt", 'r')
lines = f.readlines()
for line in lines:
    print(line)
f.close()
```

**read 함수 사용하기**

```python
f = open("python.txt", 'r')
data = f.read()
print(data)
f.close()
```

**파일에 새로운 내용 추가하기**
