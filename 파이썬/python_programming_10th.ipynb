**클래스 변수와 인스턴스 변수의 개념**

```python
class Stock:
    market = "kospi"
    composite_stock_price = 2000
    def check_market(self):
        print(self.market)
        Stock.composite_stock_price += 1

s1 = Stock()
s2 = Stock()

s1.market = "kosdaq"

s1.check_market()
s2.check_market()

Stock.market = "DowJones"
s1.check_market()
s2.check_market()
        
```

**계산기 클래스 만들기**

```python
class FourCal:
    num = 0
    def setdata(self, first, second):
        FourCal.num += 1
        self.first = first
        self.second = second
    def add(self):
        result = self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
```

```python
a = FourCal()
b = FourCal()
a.setdata(4, 2)
b.setdata(3, 8)

print("a.add() = ",a.add())
print("a.mul() = ",a.mul())
print("a.sub() = ",a.sub())
print("a.div() = ",a.div())
print("b.add() = ",b.add())
print("b.mul() = ",b.mul())
print("b.sub() = ",b.sub())
print("b.div() = ",b.div())
```
