1.12 여러조건으로 색인하기

```python
# 상권업종소분류명이 '약국'인 것과 
# '시도명'이 '서울특별시인 데이타
# and == & , or == |


df_seoul_drug = df[(df['상권업종소분류명'] == '약국') & (df['시도명'] == '서울특별시')]
print(df_seoul_drug.shape)
df_seoul_drug
# 연산자 우선순위때문에 괄호 판다스에서는 and 가 아니라 &를 사용
```

1.12.3 텍스트 데이터 색인하기

```python
#색인하기전에 상홍명에 종합병원이 아닌데이타
# loc를 쓸때는 데이타프레임형식에서 추가로 칼럼하나만 추가
df_seoul_hospital[~df_seoul_hospital['상호명'].str.contains('종합병원')]
df_seoul_hospital.loc[~df_seoul_hospital['상호명'].str.contains('종합병원'),'상호명'].unique()
```

```python
df_seoul_hospital[df_seoul_hospital['상호명'].str.contains('꽃배달')]
```

```python
df_seoul_hospital[df_seoul_hospital['상호명'].str.contains('의료기')]
```

```python
#꽃배달 의료기 장례식장 상담소 어린이집은 종합병원이 아니기에 전처리에서 제외
# 제외할 데이터의 인덱스만 drop_row에 담아주고 list형태로 변환
drop_row = df_seoul_hospital[df_seoul_hospital['상호명'].str.contains('꽃배달|의료기|장례식장|상담소|어린이집')].index
drop_row = drop_row.tolist()
print(len(drop_row))
drop_row
```

```python
drop_row2 = df_seoul_hospital[df_seoul_hospital['상호명'].str.endswith('의원')].index
drop_row2 = drop_row2.tolist()
print(len(drop_row2))
drop_row2
```

```python
drop_row = drop_row + drop_row2
len(drop_row)
```

```python
# 해당 셀을 제거하고 삭제전과 후의 갯수를 비교 
print(len(df_seoul_hospital))
df_seoul_hospital = df_seoul_hospital.drop(drop_row, axis=0)
print(len(df_seoul_hospital))
```

1.13 Folium으로 지도 활용하기

```python
import folium
df_seoul_hospital['위도'].mean()
df_seoul_hospital['경도'].mean()
# folium.Map(location=[45.5236, -122.6750])
```

```python
map = folium.Map(location=[df_seoul_hospital['위도'].mean(),
                     df_seoul_hospital['경도'].mean()],zoom_start=11)
for n in df_seoul_hospital.index:
    name = df_seoul_hospital.loc[n,'상호명']
    address = df_seoul_hospital.loc[n,'도로명주소']
    popup_content = f"<div style='width: 200px;'><b>{name}</b><br>{address}</div>"
    location = [df_seoul_hospital.loc[n,'위도'], df_seoul_hospital.loc[n,'경도']]
    folium.Marker(location, popup=popup_content).add_to(map)
map
```
