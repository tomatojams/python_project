###📌Q1. 여러분은 파이썬을 통해 설문조사 문항의 응답내역을 분석하게 되었습니다.
 문항별 응답내용에는 하나의 응답만 할 수 있는 single choice 문제와 여러 응답을 선택할 수 있는 multiple choice 문제가 있습니다. 2개를 구분하기 위해 single choice 문항 번호에 "_"를 표기하지 않기로 했습니다.

문항별 응답내역이 'question'에 담겨 있을 때, 조건문과 반복문을 사용하여 아래와 같은 결과가 출력되도록 코드를 작성해보세요.


✅⚡  미션분석과 개선사항

0. 응답내역(문제의 question)에서 싱글문항만을 걸러내는 조건(single_question_list())은 너무 간단하므로, 약간의 시나리오를 더해서 문제를 풀어봅니다.

1. 본인이 응답내역을 분석한다면 설문응답자의 raw데이타에서 분석을 시작할 것입니다. 이러한 비교적 정리되지않은 자료형에서 각 질문에 대한 응답내역(문제의 question)을 추출하는 데이타를 생성합니다. 그렇게 가공한 후에 대답한 싱글문항만을 출력하는 함수또한 따로생성하게 됩니다.

2.  설문조사의 문항 제작단계에서, 문항별 질문수의 문서가 이미 있을 것이므로, 그것을 가정해 질문별 다문항의 수가 기입된 자료형을 만듭니다.

3. ⚡ (추가기능) 다문항의 세부문항에서도 응답자가 대답을 건너뛴 문항이 있다고 가정하고 그부분을 건너뛰고 출력하는 기능을 만들어봅니다.
다문항중 Q6-6번을 기입하지 않았다고 가정합니다.

4. (목표) 따라서 명확하게 어떠한 문제에 응답했는지, 한눈에 알아보기 힘든 설문응답 raw 데이타에 대해서, 설문제작단계의 자료와 비교해, 대답한 문항들을 출력하는 함수를 만드는 시나리오를 상정하고 제작했습니다.

````python
#Q1

# 설문전 만들어진 다문항에 대한 자료형
length_of_question = {'Q1': 'single', 'Q2': 'single', 'Q3': 'single', 'Q4': 'single',
                      'Q5': 'single', 'Q6': 12, 'Q7': 7, 'Q8': 'single',
                      'Q9': 'single', 'Q10': 3}

# 설문과정에서 얻은 raw 데이타를 바탕으로하는 응답자료
surveyed_data = {'Q1': None, 'Q2': 'Yes', 'Q3': 'No', 'Q4': 'Yes', 'Q5': 'Yes',
                 'Q6': [1, 2, 1, 4, 2, None, 3, 5, 1, 2, 4, 3],
                 'Q7': [2, 3, 1, 2, 3, 5, 4], 'Q8': 'single', 'Q9': 'single', 'Q10': [3, 1, 1]}


def survey_to_answered_list(surveyed_data: dict) -> list:
    """
    복잡하게 응답된 질문들의 딕셔너리를, 각 문항에 대한 단순히 응답여부를 확인하는 목록으로 변환합니다.

    Args:
        surveyed_data (dict): 응답된 질문들을 포함하는 딕셔너리입니다. 여기서 키는 질문 식별자이며,
                                 값은 응답되지 않은 질문에 대해서는 'None'이고, 그렇지 않으면 응답 목록입니다.

    Returns:
        list: 응답된 질문들의 목록입니다 (멀티문항의 경우 하위 질문도 포함됩니다).
    """

    answered_list = []

    for index, answer in surveyed_data.items():
        if answer is None:
            continue
        if length_of_question[index] == 'single':
            answered_list.append(index)
        else:
            for sub_idx, sub_ans in enumerate(answer):
                if sub_ans is not None:
                    answered_list.append(f'{index}_{sub_idx + 1}')

    return answered_list


def single_question_list(answered_list: list) -> list:
    """
    주어진 질문 목록에서 멀티 질문을 제외하고 단일 질문 목록을 반환합니다.

    Args:
        question_list (list): 질문 목록

    Returns:
        list: 단일 질문 목록
    """
    single_question = []
    for question in answered_list:
        if '_' not in question:
            single_question.append(question)
    return single_question


# 예시 사용:
print('question =', survey_to_answered_list(surveyed_data))
print('answered single-question =', single_question_list(survey_to_answered_list(surveyed_data)))
````

````python
#Q2-1 pivot_table을 활용한 풀이

import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")

# 도표를 2000년부터 2010년까지로제한
filtered_df = df[(df['Year'] >= 2000) & (df['Year'] <= 2010)]

# pivot_table을 활용해서 도표생성
pd.pivot_table(filtered_df, values='Life_Expectancy', index='Year', columns='Country')
````

````python
#Q2-2 groupby를 활용한 풀이
import pandas as pd

# 데이터 로드
df2 = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")

# groupby와 unstack를 사용하여 인덱스를 만들고 평균 수명 계산
result = df2[df2['Year'].between(2000, 2010)].groupby(['Year', 'Country'])[
    'Life_Expectancy'].mean().unstack()

# 결과 출력
print(result)

````
