4. (ëª©í‘œ) ë”°ë¼ì„œ ëª…í™•í•˜ê²Œ ì–´ë– í•œ ë¬¸ì œì— ì‘ë‹µí–ˆëŠ”ì§€, í•œëˆˆì— ì•Œì•„ë³´ê¸° í˜ë“  ì„¤ë¬¸ì‘ë‹µ raw ë°ì´íƒ€ì— ëŒ€í•´ì„œ, ì„¤ë¬¸ì œì‘ë‹¨ê³„ì˜ ìë£Œì™€ ë¹„êµí•´, ëŒ€ë‹µí•œ ë¬¸í•­ë“¤ì„ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“œëŠ” ì‹œë‚˜ë¦¬ì˜¤ë¥¼ ìƒì •í•˜ê³  ì œì‘í–ˆìŠµë‹ˆë‹¤.

````python
#Q1

# ì„¤ë¬¸ì „ ë§Œë“¤ì–´ì§„ ë‹¤ë¬¸í•­ì— ëŒ€í•œ ìë£Œí˜•
length_of_question = {'Q1': 'single', 'Q2': 'single', 'Q3': 'single', 'Q4': 'single',
                      'Q5': 'single', 'Q6': 12, 'Q7': 7, 'Q8': 'single',
                      'Q9': 'single', 'Q10': 3}

# ì„¤ë¬¸ê³¼ì •ì—ì„œ ì–»ì€ raw ë°ì´íƒ€ë¥¼ ë°”íƒ•ìœ¼ë¡œí•˜ëŠ” ì‘ë‹µìë£Œ
surveyed_data = {'Q1': None, 'Q2': 'Yes', 'Q3': 'No', 'Q4': 'Yes', 'Q5': 'Yes',
                 'Q6': [1, 2, 1, 4, 2, None, 3, 5, 1, 2, 4, 3],
                 'Q7': [2, 3, 1, 2, 3, 5, 4], 'Q8': 'single', 'Q9': 'single', 'Q10': [3, 1, 1]}


def survey_to_answered_list(surveyed_data: dict) -> list:
    """
    ë³µì¡í•˜ê²Œ ì‘ë‹µëœ ì§ˆë¬¸ë“¤ì˜ ë”•ì…”ë„ˆë¦¬ë¥¼, ê° ë¬¸í•­ì— ëŒ€í•œ ë‹¨ìˆœíˆ ì‘ë‹µì—¬ë¶€ë¥¼ í™•ì¸í•˜ëŠ” ëª©ë¡ìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.

    Args:
        surveyed_data (dict): ì‘ë‹µëœ ì§ˆë¬¸ë“¤ì„ í¬í•¨í•˜ëŠ” ë”•ì…”ë„ˆë¦¬ì…ë‹ˆë‹¤. ì—¬ê¸°ì„œ í‚¤ëŠ” ì§ˆë¬¸ ì‹ë³„ìì´ë©°,
                                 ê°’ì€ ì‘ë‹µë˜ì§€ ì•Šì€ ì§ˆë¬¸ì— ëŒ€í•´ì„œëŠ” 'None'ì´ê³ , ê·¸ë ‡ì§€ ì•Šìœ¼ë©´ ì‘ë‹µ ëª©ë¡ì…ë‹ˆë‹¤.

    Returns:
        list: ì‘ë‹µëœ ì§ˆë¬¸ë“¤ì˜ ëª©ë¡ì…ë‹ˆë‹¤ (ë©€í‹°ë¬¸í•­ì˜ ê²½ìš° í•˜ìœ„ ì§ˆë¬¸ë„ í¬í•¨ë©ë‹ˆë‹¤).
    """

    answered_list = []

    for index, answer in surveyed_data.items():
        if answer is None:
            continue
        if length_of_question[index] == 'single':
            answered_list.append(index)
        else:
            for sub_idx, sub_ans in enumerate(answer):
                if sub_ans is not None:
                    answered_list.append(f'{index}_{sub_idx + 1}')

    return answered_list


def single_question_list(answered_list: list) -> list:
    """
    ì£¼ì–´ì§„ ì§ˆë¬¸ ëª©ë¡ì—ì„œ ë©€í‹° ì§ˆë¬¸ì„ ì œì™¸í•˜ê³  ë‹¨ì¼ ì§ˆë¬¸ ëª©ë¡ì„ ë°˜í™˜í•©ë‹ˆë‹¤.

    Args:
        question_list (list): ì§ˆë¬¸ ëª©ë¡

    Returns:
        list: ë‹¨ì¼ ì§ˆë¬¸ ëª©ë¡
    """
    single_question = []
    for question in answered_list:
        if '_' not in question:
            single_question.append(question)
    return single_question


# ì˜ˆì‹œ ì‚¬ìš©:
print('question =', survey_to_answered_list(surveyed_data))
print('answered single-question =', single_question_list(survey_to_answered_list(surveyed_data)))
````

````python
#Q2-2 groupbyë¥¼ í™œìš©í•œ í’€ì´
import pandas as pd

# ë°ì´í„° ë¡œë“œ
df2 = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")

# groupbyì™€ unstackë¥¼ ì‚¬ìš©í•˜ì—¬ ì¸ë±ìŠ¤ë¥¼ ë§Œë“¤ê³  í‰ê·  ìˆ˜ëª… ê³„ì‚°
result = df2[df2['Year'].between(2000, 2010)].groupby(['Year', 'Country'])[
    'Life_Expectancy'].mean().unstack()

# ê²°ê³¼ ì¶œë ¥
print(result)

````
