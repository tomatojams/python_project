```python
def isValid(str):
    string = [ i for i in str] #리스트 생성
    for _ in range(len(string)//2):

        for l,num in enumerate(string):
            if ord(num) == ord(string[0]) +1 or ord(num) == ord(string[0])+2:   #아스키 코드로 커플링검사 +1,+2 로 순서확인
                del string[l], string[0]
                break   #처음과 같은거 하나만 찾고 빠져나옴

        if len(string)==0:  #모두 삭제되면 유효
            return True

    return False  #모두 삭제되지 않았을때 for문 빠져나옴
```

```python
def solution(s):
    string = [ord(i) for i in s]
    loop = len(string)//2
    for _ in range(loop):  #  리스트크기의 절반만 실행

        if string[0] + 1 in string:
            string.remove(41)
            del string[0]
        elif (string[0] +2) in string:
            string.remove(string[0]+2)
            del string[0]

    if len(string) == 0:  # 모두 삭제되면 유효
        return True
    return False
```

```python
def isValid(s):

    opening = "{[("
    closing = "}])"

    validChar = ['(',')','}','{','[',']']

    stack1 = []
    stack2 = []
    stack3 = []

    for char in s:
        if char in validChar:
            if char in opening:
                if char == '{':
                    stack1.append(char)
                elif char == '[':
                    stack2.append(char)
                elif char == '(':
                    stack3.append(char)
            elif char in closing:
                if char == '}' and stack1:
                    stack1.pop()
                elif char == ']' and stack2:
                    stack2.pop()
                elif char == ')' and stack3:
                    stack3.pop()
                else:
                    return False
        else:
            return False
    return len(stack1) == 0 and len(stack2) == 0 and len(stack3) == 0
```

```python
def isValid(s):

    opening = ['(', '[', '{']
    closing = [')', ']', '}']
    stack = []
    result = True

    for i in s:
        if i in opening:
            stack.append(i)
        elif i in closing:
            if stack == []:
                result = False
                break
            elif stack[-1] in opening:
                if opening[closing.index(i)] in stack:
                    stack.remove(opening[closing.index(i)])
                else:
                    result = False
                    break
            else:
                result = False
                break

    if stack != []:
        result = False

    return result
```
