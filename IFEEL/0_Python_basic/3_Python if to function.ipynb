## Python if

### if문을 이용해서 판별하기
* 숫자를 입력했을 때, 그 숫자가 짝수/홀수를 판별해라.


````python
# 숫자를 입력했을 때,
num = input(" ")

if int(num) % 2 == 0:
    print("짝수")
else:
    print("홀수")

if int(num) % 2 == 1:
    print("홀수")
else:
    print("짝수")
````

### if문을 이용한 점수 분류
* 입력된 점수를 100점부터 20점 단위로 A, B, C 학점을 부여해주세요.

````python
score = input("score : ")
score = int(score)

if 80 < score <=100:
    print("A")
elif 60 < score <= 80:
    print("B")
elif 40 < score <= 60:
    print("C")
````

## Python Loop

### for 문을 이용한 List 출력
* 주어진 List를 for문 이용해 출력해보세요.
```
city = ["서울", "인천", "경기", "강원", "제주"]
```

````python
city = ["서울", "인천", "경기", "강원", "제주"]

for c in city:
    print(c)
````

---

### for문과 range를 이용한 출력 조절
* 1부터 50까지 숫자 중 5의 배수를 출력해보세요.

````python
for num in range(5, 51, 5):
    print(num)
````

---

### for문으로 출력 조절
* 아래 출력을 for 문을 이용해서 구현해보세요.
```python
print(10)
print("-------")
print(20)
print("-------")
print(30)
print("-------")
```

````python
for i in range(10,40,10):
    print(i)
    print("-------")

for i in range(0,30,10):
    print(i + 10)
    print("-------")
````

---

### 구구단 출력하기
* for문을 이용해서 구구단 2단을 출력해보세요.

````python
for j in range(2, 10):
    for i in range(1, 10): # 1~9
        print(j, "x", i, " = ", j * i)
    print("----")
````
