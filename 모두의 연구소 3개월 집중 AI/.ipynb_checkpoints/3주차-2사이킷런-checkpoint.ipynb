### 원핫(one-hot)인코딩

```python
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
df
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False) # 어레이형태로 받기위함
cat = ohe.fit_transform(df[['원산지']]) # onehot은 데이타프레임형태로 넣어야함
cat
```

```python
ohe.categories_
```

```python
#피쳐(칼럼)과 카테고리를 붙임
ohe.get_feature_names_out()
```

```python
# 데이타프레읨으로 변환
df_cat = pd.DataFrame(cat, columns = ohe.get_feature_names_out() )
```

```python
df =pd.concat([df,df_cat], axis= 1)
df
```

```python
df = df.drop(['원산지'], axis = 1)
df
```

```python
# 여러개 원핫인코딩
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
df
```

```python
from sklearn.preprocessing import OneHotEncoder
cols = df.select_dtypes(include= 'object').columns
ohe = OneHotEncoder(sparse_output=False) # 어레이형태로 받기위함
cat = ohe.fit_transform(df[cols]) # onehot은 데이타프레임형태로 넣어야함
cat
```

```python
ohe.get_feature_names_out()
```

```python
df_cat = pd.DataFrame(cat,columns= ohe.get_feature_names_out())
df_cat
```

```python
df = pd.concat([df,df_cat],axis=1)
df
```

```python
df = df.drop(cols, axis=1)
df
```

## 데이타전처리 :수치형 데이터

### 민맥스 (MinMaxSCaler)
