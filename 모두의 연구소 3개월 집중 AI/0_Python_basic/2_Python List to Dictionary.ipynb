### List 선언하기
* 아래 데이터들을 이용해서 List를 선언해보세요.
```
서울, 인천, 경기, 충청, 전라, 경상, 강원
```

````python
city = ["서울", "인천", "경기", "충청", "전라", "경상", "강원"]
print(city)
````

---

## Python Tuple

### Tuple 생성하기
* 다음 데이터를 이용해서 튜플을 생성해보세요.
```python
사과, 체리, 바나나, 참외
```

````python
fruits = ('사과', '체리', '바나나', '참외')
print(fruits)
````

---


### Tuple을 List로 변환
* 아래 데이터를 Tuple로 만들고 리스트로 변환해보세요.
```python
fruits = ('사과', '체리', '바나나', '참외')
```

````python
fruits = ('사과', '체리', '바나나', '참외')
print(fruits)

list_fruits = list(fruits)
print(list_fruits)
````

---

### Unpacking Tuple
* 아래 튜플을 unpacking해서 출력해보세요.
```python
fruits = ('사과', '체리', '바나나', '참외')
```

````python
fruits = ('사과', '체리', '바나나', '참외')
print(fruits)

a, b, c, d = fruits
print(a, b)
print(c, d)
````

## Python Dictionary

### 데이터를 가지고 있지 않은 Dictionary
* 데이터 없이 선언된 dictionary 변수를 생성해보세요.

````python
no_data = {"김밥":1000, "라면":2000}
print(no_data)

no_data["우동"] = 2500
print(no_data)
````

---

### Dictionary 데이터 추가
* key, value로 이뤄진 데이터를 추가해보세요.
```python
국어 85, 수학 80, 영어 75
```

````python
score = {"국어":85, "수학":80, "영어":75}
print(score)
````

---

### 기존 Dictionary에 새로운 데이터 추가
* 위에 선언한 score 변수에 아래 key, value를 추가해보세요.
```python
사회 80, 과학 85
```

````python
score["사회"] = 80
score["과학"] = 85
print(score)
````

---

### Dictionary 데이터 삭제
* 위 score 값 중 사회 값을 삭제해보세요.

````python
del score["사회"]
print(score)
````

---

### Dictionary 값 업데이트
* score 데이터 중 수학의 value를 85로 변경해보세요.

````python
score["수학"] = 85
print(score)
````

---

### Dictionary 에서 key 와 value 값 출력
* 주어진 dictionary 데이터가 있을때, key, value 값을 따로 출력해보세요.

````python
print(score.keys())
print(score.values())
````
