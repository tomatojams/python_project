```python
# 데이터 관련 라이브러리
import numpy as np

# classification 관련 라이브러리
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Resampling 관련 라이브러리
from sklearn.model_selection import KFold  # K-fold cross validation
from sklearn.utils import resample  # bootstrap

# 임의로 데이터를 섞기위한 라이브러리(random shuffling)
from sklearn.utils import shuffle

# 붓꽃(Iris) 데이터 로딩
from sklearn.datasets import load_iris
```

K-fold cross validation

```python
X=["a","b","c","d"]
kf = KFold(n_splits=2)
for train, test in kf.split(X):   # X를 나눠라 두번을 나누는데 서로세팅이 바뀜
    print("%s %s" %(train, test))
    print(X[train[0]], X[train[1]], X[test[0]], X[test[1]]) 
```

붓꽃 데이타예제(K-fold cross validation)

```python
X_iris, y_iris = load_iris(return_X_y = True)
```

```python
X_iris
```

```python
y_iris
```

```python
X_iris,y_iris = shuffle(X_iris, y_iris) #서플로 섞음
```

```python
# 5-fold cross-validation

kf = KFold(n_splits=5)
idx = 1
for train, val in kf.split(X_iris): #X_iris를 나눠라
    print("Fold: #",idx)

    ## K-Fold 결과확인
    print("%s %s" %(train, val)) #인덱스값만 리턴
    print(y_iris[train])
    print(y_iris[val])

    #LDA
    #training
    clf = LinearDiscriminantAnalysis().fit(X_iris[train], y_iris[train]) #Kfold로 나눈 train만 학습에 사용

    #클래스예측
    print("true label:", y_iris[val])
    print("pred label:", clf.predict(X_iris[val]))

    #accuracy
    print("accuracy:", clf.score(X_iris[val], y_iris[val])) #정확도평가

    print("---------------------------------------")
    idx+=1
                                           
```
