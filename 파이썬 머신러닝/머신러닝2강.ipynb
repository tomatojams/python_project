```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# 데이타생성
x = np.linspace(0, 10, 100)  #구간 시작점, 구간 끝점, 구간 내 숫자 개수 구간 내에 숫자를 균일한 간격으로 촘촘하게 채울 수 있는 np.linspace 
#간격이 촘촘할수록 그래프를 그리는 상황 등에서 실제 그래프 개형과 더 가까워진다는 장점이 있지만, 어레이 사이즈가 커져서 다루기 힘들어질 수 있다는 단점이 있습니다.
y = 4 + 2 * np.sin(2 * x)

#Figure 객체를 생성하고 
#해당 Figure 객체에 여러 개의 AxesSubplot 객체를 생성하는 두 가지 작업을 한번에 하고자 할 때는 다음과 같이 plt.subplots를 사용하면 됩니다.

fig, ax = plt.subplots()

ax.plot(x, y, linewidth=5.0)

ax.set(xlim=(0, 8), xticks=np.arange(2, 6),  #lim 눈끔의 범위가 0~8 # tick - 눈끔 하나하나 1~7(마지막전)
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.grid()
plt.show()
```

```python
# 데이터 로딩
#from google.colab import drive
#drive.mount('/content/drive')

ad = pd.read_csv("data/Advertising.csv")
```

```python
# 데이터 생성
tv = ad["TV"]
radio = ad["Radio"]
news = ad["Newspaper"]
sales = ad["Sales"]

#plot
fig, ax = plt.subplots()

ax.plot(tv, linewidth=1.0)
ax.plot(radio, linewidth=1.0)
ax.plot(news, linewidth=1.0)
ax.plot(sales, linewidth=1.0)

plt.grid()
plt.show()
```

**scatter() 예제**


```python
np.random.seed(42)
x = 4 + np.random.normal(0, 2, 24)  #정규분포 평균 0 표준편차 2 의 24개 데이타 생성
y = 4 + np.random.normal(0, 2, len(x))

sizes = np.random.uniform(15, 80, len(x))  # 균등분포 15~80사이에서 24개를 생성
colors = np.random.uniform(15, 80, len(x))

fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.grid()
plt.show()

```

**matplotlib 통계관련 plot**
hist()

```python
#데이타생성
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

#plot:
fig, ax = plt.subplots()

ax.hist(x, bins=20, linewidth=3, edgecolor="green")  # bins 가로칸수, 선두께 엣지칼라=외곽선

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 56), yticks=np.linspace(0, 56, 9))

plt.grid()
plt.show()
```

**boxplot()예제**

```python
#데이타
np.random.seed(42)
D = np.random.normal((3, 5, 10), (1.25, 1.00, 1.25), (100, 3))  #첫!) 세개모두 평균, 두번째 ()  세번째 데이타개수

#plot
fig, ax = plt.subplots()
ax.boxplot(D)

plt.grid()
plt.show()
```
