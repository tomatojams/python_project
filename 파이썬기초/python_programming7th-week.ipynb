# while 문

```python
treeHit = 0
while treeHit <10:
    treeHit +=1
    print("나무를 %d번 찍었습니다" %treeHit)
    if treeHit == 10:
        print("나무 넘어갑니다")
```

```python
treeHit = 0
while treeHit < 10:
    treeHit = treeHit + 1
    print("나무를 %d 번 찍었습니다."% treeHit)
    if treeHit == 10:
        print("나무 넘어 갑니다.")
```

**break 사용하기**

```python
coffee = 10
money = 300
while money:
    print("돈을 받았으니 커피를 줍니다.")
    coffee = coffee  -1
    print("남은 커피양은 %d개입니다." % coffee)
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다")
        break
```

**continue 사용하기**

```python
a= 0. #홀수만 걸러내기
while a <10:
    a = a+1
    if a % 2 ==0:
        continue
    print(a)
```

# for 문

**리스트를 이용한 for 반복문**

```python
test_list = ['one', 'two', 'three'] #리스트를 사용한 반복문
for i in test_list:
    print(i)
```

**range() 함수의 사용**

```python
a = range(10)
a
```

```python
a = range(1,11)
a
```

**튜플을 이용한 for 문**

```python
a = (1,2,3)

for i in a:
    print(i)
```

**문자열을 이용한 for 문**

```python
word = "I love python" #리스트, 튜플, 문자열로 for문을 사용가능

for i in word:
    print(i)
```

**이중 for 문**

```python
for i in range(2,10):        # ①번 for문
    for j in range(1, 10):   # ②번 for문
        print(i*j, end=" ") 
    print("")
```

```python
a = [1,2,3,4]
result =[]
for num in a:
    result.append(num*3)
print(result)
```

```python
a = [1,2,3,4]
result = [num*3 for num in a if(num%2 ==0)]
print(result)
```

**내포 리스트**

```python
result = [x*y for x in range(2,10) for y in range(1,10)]
print(result)
```

# 학생들의 성적 처리 하기

```python
score = [[23,90,20,90],[25,50,20,10],[15,60,5,70],[0,30,20,20],[0,10,0,30],[5,70,5,100]]

grade = []

for i in score:

    sum = 0
    for j in i:
        sum +=j
    grade.append(sum)

print(grade)
```
